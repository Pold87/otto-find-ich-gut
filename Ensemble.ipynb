## Polish submission file

The following function takes the sum of all probabilities below 0.10 and add them to the maximum probability value. Afterwards, the probabilities below 0.10 are set to 0.

```python
new_df = pd.DataFrame(columns = raw_submission.columns)

for i, row in raw_submission.iterrows():
    probs = row[1:]
    row_max = probs.max()
    idx_max = probs.idxmax()
    below_10_mask = probs < 0.10
    below_10 = probs[below_10_mask]
    sum_below_10 = below_10.sum()
    
    probs[idx_max] = row_max + sum_below_10
    probs[below_10_mask] = 0
    
    new_probs = row
    new_probs[1:] = probs
    
    raw_submission.iloc[i] = new_probs
    
    
    
```
