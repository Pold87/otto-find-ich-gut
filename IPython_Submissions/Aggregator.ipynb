```python
from hyperopt import fmin, tpe, hp, Trials, STATUS_OK
import functools

space = {'keras1' : hp.uniform('keras1', 0, 1),
         'keras2' : hp.uniform('keras2', 0, 1),
         'h2o1' : hp.uniform('h2o1', 0, 1),
         'h2o2' : hp.uniform('h2o2', 0, 1),
         'lasagne' : hp.uniform('lasagne', 0, 1),
         'xgb1' : hp.uniform('xgb1', 0, 1),
         'xgb2' : hp.uniform('xgb2', 0, 1),
#         'delta' : hp.uniform('delta', 0, 0.1),
#         'factor' : hp.uniform('factor', 0, 0.1)
        }
```

```python
sample_sub = "../submissions/sampleSubmission.csv"
sample_sub_df = pd.read_csv(sample_sub)

def normalize(row, epsilon=1e-15):
    
    row = row / np.sum(row)
    row = np.maximum(epsilon, row)
    row = np.minimum(1 - epsilon, row)
    
    return row
    
def logloss_mc(y_true, y_probs):
    
    # Normalize probability data frame
    y_probs = y_probs.apply(normalize, axis=1)
        
    log_vals = []
        
    for i, y in enumerate(y_true):
        c = int(y.split("_")[1])
        log_vals.append(- np.log(y_probs.iloc[i,c - 1]))
        
    return np.mean(log_vals)
```

```python
from math import erf

def postprocessor_helper(x, delta, factor):
    if x < delta:
        return x * factor
    else:
        return x
    
def postprocessor(df, delta, factor):
    return df.applymap(functools.partial(postprocessor_helper,
                                         delta = delta,
                                         factor = factor))


def postprocessor_platt_helper(x, a, b):
    return x + a * (b - x)
    
def postprocessor_platt(df, a, b):
    return df.applymap(functools.partial(postprocessor_platt_helper,
                                         a = a,
                                         b = b))

def postprocessor_no_args_helper(x):
    return erf(x)
    #return (1 / (1 + np.exp(- 4.2 * x))) - 0.5
    
def postprocessor_no_args(df):
    return df.applymap(postprocessor_no_args_helper)
```

```python
df_holdout = pd.read_csv("../data/test25.csv")
y_valid = df_holdout.target

sub_keras1_X = pd.read_csv("../IPython_Submissions/keras/keras-otto-final-aggregated-1.csv").iloc[:, 1:]
sub_keras2_X = pd.read_csv("../IPython_Submissions/keras/keras-otto-cross_25.csv").iloc[:, 1:]

sub_h2o1_X = pd.read_csv("../h2onewcrossvalidation_100_epochs.csv").iloc[:, 1:]
sub_h2o2_X = pd.read_csv("../h2onewcrossvalidation.csv").iloc[:, 1:]

sub_lasagne_X = pd.read_csv("lasagne-otto-final-25.csv").iloc[:, 1:]

sub_xgb1_X = pd.read_csv("../IPython_Submissions/xgboostpython_25.csv").iloc[:, 1:]
sub_xgb2_X = pd.read_csv("../IPython_Submissions/xgboostpython_25_2.csv").iloc[:, 1:]

```

```python
def objective(params):

    sub = params['keras1'] * sub_keras1_X + \
          params['keras2'] * sub_keras2_X + \
          params['h2o1'] * sub_h2o1_X + \
          params['h2o2'] * sub_h2o2_X + \
          params['lasagne'] * sub_lasagne_X + \
          params['xgb1'] * sub_xgb1_X + \
          params['xgb2'] * sub_xgb2_X
    
    factors = params['keras1'] + params['keras2'] + params['h2o1'] + params['h2o2'] + params['lasagne'] + params['xgb1'] + params['xgb2']
    
    sub = sub / factors
    
    # Postprocessing
    #sub = postprocessor(sub,
    #                    delta = params['delta'],
    #                    factor = params['factor'])

    #sub = postprocessor_platt(sub,
    #                    a = params['delta'],
    #                    b = params['factor'])

    sub = postprocessor_no_args(sub)
    
    ll = logloss_mc(y_valid, sub)
    print(ll)
    return {'loss' : ll,
            'status' : STATUS_OK}


trials = Trials()

best = fmin(fn=objective,
    space=space,
    algo=tpe.suggest,
    max_evals=300,
    trials=trials)

print best
```
